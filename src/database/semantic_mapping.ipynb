```python
tables = client.get_tables(database_id=database_id)

for table in tables:
    print(table.id, table.name)
```

```python
for table in tables:
    print(f"\nTABLE: {table.name}")

    details = client.get_table(
        database_id=database_id,
        table_id=table.id
    )

    for col in details.columns:
        print(col.id, col.name)
```
