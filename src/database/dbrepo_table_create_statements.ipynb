### accident

```python
table_accident = {
    "name": "accident",
    #"description": "The table is storing general information about road traffic accidents",
    "columns": [
        {
            "name": "reference_number",
            "type": "varchar",
            "size": 20,
            "null_allowed": False,
            "description": "The accident reference number"
        },
        {
            "name": "easting",
            "type": "int",
            "null_allowed": False,
            "description": "The easting coordinate of the accident location"
        },
        {
            "name": "northing",
            "type": "int",
            "null_allowed": False,
            "description": "The northing coordinate of the accident location"
        },
        {
            "name": "number_of_vehicles",
            "type": "int",
            "null_allowed": False,
            "description": "The number of vehicles involved in the accident"
        },
        {
            "name": "date",
            "type": "date",
            "null_allowed": False,
            "description": "The date of the accident"
        },
        {
            "name": "time",
            "type": "int",
            "null_allowed": False,
            "description": "The time of the accident"
        },
        {
            "name": "road_class_id",
            "type": "int",
            "null_allowed": False,
            "description": "Reference to the road class"
        },
        {
            "name": "road_surface_id",
            "type": "int",
            "null_allowed": False,
            "description": "Reference to the road surface condition"
        },
        {
            "name": "lighting_id",
            "type": "int",
            "null_allowed": False,
            "description": "Reference to the lighting condition"
        },
        {
            "name": "weather_id",
            "type": "int",
            "null_allowed": False,
            "description": "Reference to the weather condition"
        }
    ],
    "constraints": {
        "primary_key": ["reference_number"],
        "foreign_keys": [
            {
                "columns": ["road_class_id"],
                "referenced_table": "road_class",
                "referenced_columns": ["road_class_id"]
            },
            {
                "columns": ["road_surface_id"],
                "referenced_table": "road_surface",
                "referenced_columns": ["road_surface_id"]
            },
            {
                "columns": ["lighting_id"],
                "referenced_table": "lighting",
                "referenced_columns": ["lighting_id"]
            },
            {
                "columns": ["weather_id"],
                "referenced_table": "weather",
                "referenced_columns": ["weather_id"]
            }
        ],
        "uniques": [],
        "checks": [],
    },
    "is_public": False,
    "is_schema_public": False
}
```

```python
import json

response = requests.post(
    url_tables, 
    json = table_accident,
    auth = HTTPBasicAuth(username, password)
) 


#print(response.request.headers)
#print(response.request.body)

print(response.status_code)
print(response.text)
print(response.headers)
```
